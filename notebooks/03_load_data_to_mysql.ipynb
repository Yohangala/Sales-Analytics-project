```python
import pandas as pd
from sqlalchemy import create_engine

df = pd.read_csv("../Data/cleaned/global_superstore_cleaned.csv")
df.shape
```

```python
engine = create_engine(
    "mysql+pymysql://root:12345@localhost:3306/sales_analytics"
)
```

```python
df.to_sql(
    name="orders",
    con=engine,
    if_exists="replace",
    index=False,
    chunksize=1000,
    method="multi"
)
```
